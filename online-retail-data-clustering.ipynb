Data Cleaning


```python
cleaned_df = df.copy()
```

astype("str"): converts the Invoice column to strings.

str.match(r"^\d{6}$"): True for values that are exactly six digits (no letters/spaces).

cleaned_df[mask]: filters the DataFrame to those rows.

```python
cleaned_df["Invoice"] = cleaned_df["Invoice"].astype("str")

mask = (
    cleaned_df["Invoice"].str.match("^\\d{6}$") == True
)

cleaned_df = cleaned_df[mask]

cleaned_df
```

```python
cleaned_df["StockCode"] = cleaned_df["StockCode"].astype("str")

mask = (
    (cleaned_df["StockCode"].str.match("^\\d{5}$") == True)
    | (cleaned_df["StockCode"].str.match("^\\d{5}[a-zA-Z]+$") == True)
    | (cleaned_df["StockCode"].str.match("^PADS$") == True)
)

cleaned_df = cleaned_df[mask]

cleaned_df
```

This drops any rows where Customer ID is missing. dropna(subset=["Customer ID"], inplace=True) means “remove rows with NaN in the Customer ID column, and update cleaned_df in place.”

```python
cleaned_df.dropna(subset=["Customer ID"], inplace=True)
```

Feature Engineering

```python
cleaned_df["SalesLineTotal"] = cleaned_df["Quantity"] * cleaned_df["Price"]
cleaned_df
```

aggregated_df is summarizing each customer:

groupby(by="Customer ID", as_index=False): group rows by customer.

MonetaryValue=("SalesLineTotal", "sum"): total spend per customer.

Frequency=("Invoice", "nunique"): number of unique invoices per customer.

LastInvoiceDate=("InvoiceDate", "max"): most recent invoice date per customer.

```python
aggregated_df = cleaned_df.groupby(by="Customer ID", as_index=False) \
    .agg(
        MonetaryValue=("SalesLineTotal","sum"),
        Frequency=("Invoice", "nunique"),
        LastInvoiceDate=("InvoiceDate", "max")
    )

aggregated_df.head(10)
```

To check for recency

```python
max_invoice_date = aggregated_df["LastInvoiceDate"].max()
max_invoice_date
```

max_invoice_date takes the latest LastInvoiceDate in the aggregated table, then Recency calculates how many days each customer is from that latest date:

max_invoice_date = aggregated_df["LastInvoiceDate"].max(): find the most recent invoice date overall.

aggregated_df["Recency"] = (max_invoice_date - aggregated_df["LastInvoiceDate"]).dt.days: for each row, subtract that row’s last invoice date from the overall max and take the difference in days; smaller numbers mean more recent activity.

```python
max_invoice_date = aggregated_df["LastInvoiceDate"].max()
aggregated_df["Recency"] = (max_invoice_date - aggregated_df["LastInvoiceDate"]).dt.days
aggregated_df.head(5)
```

Removing outliers

```python
M_Q1 = aggregated_df["MonetaryValue"].quantile(0.25)
M_Q3 = aggregated_df["MonetaryValue"].quantile(0.75)
M_IQR = M_Q3 - M_Q1
monetary_outliers_df = aggregated_df[(aggregated_df["MonetaryValue"] > (M_Q3 + 1.5 * M_IQR)) | (aggregated_df["MonetaryValue"] < (M_Q1 - 1.5 * M_IQR))].copy()

monetary_outliers_df.describe()
```

```python
F_Q1 = aggregated_df["Frequency"].quantile(0.25)
F_Q3 = aggregated_df["Frequency"].quantile(0.75)
F_IQR = F_Q3 - F_Q1
frequency_outliers_df = aggregated_df[(aggregated_df["Frequency"] > (F_Q3 + 1.5 * F_IQR)) | (aggregated_df["Frequency"] < (F_Q1 - 1.5 * F_IQR))].copy()

frequency_outliers_df.describe()
```

```python
non_outliers_df = aggregated_df[(~aggregated_df.index.isin(monetary_outliers_df.index)) & (~aggregated_df.index.isin(frequency_outliers_df.index))] 
non_outliers_df.describe()
```
